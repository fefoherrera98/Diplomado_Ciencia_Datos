```python
import pandas as pd

class IrisDataIterator:
    def __init__(self, filename):
        self.data = pd.read_csv(filename)
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.data):
            raise StopIteration("Los datos se han agotado")
        else:
            result = self.data.iloc[self.index+1:self.index + 3]
            self.index += 1
            return result


# Prueba de la clase IrisDataIterator
iris_iterator = IrisDataIterator('iris.csv')

for batch in iris_iterator:
    print(batch)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

class DescriptiveStatistics:
    def __init__(self, data_file):
        self.data = pd.read_csv(data_file)
    
    def calculate_basic_stats(self):
        return self.data.describe()
    
    def get_variable_names(self):
        return self.data.columns.tolist()
    
    def get_variable_stats(self, variable_name):
        if variable_name not in self.data.columns:
            raise ValueError("Variable not found in the dataset.")
        
        variable_data = self.data[variable_name]
        stats = {
            'Mean': variable_data.mean(),
            'Median': variable_data.median(),
            'Standard Deviation': variable_data.std(),
            '25th Percentile': variable_data.quantile(0.25),
            '50th Percentile': variable_data.quantile(0.5),
            '75th Percentile': variable_data.quantile(0.75)
        }
        
        plt.hist(variable_data, bins='auto')
        plt.xlabel(variable_name)
        plt.ylabel('Frequency')
        plt.title('Histogram of ' + variable_name)
        plt.show()
        
        return stats


# Prueba de la clase DescriptiveStatistics
stats = DescriptiveStatistics('iris.csv')
print(stats.calculate_basic_stats())
print(stats.get_variable_names())
print(stats.get_variable_stats('sepal.length'))
```
