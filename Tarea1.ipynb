## Definition

Average is the sum of a collection of numbers divided by the count of numbers in the collection.  For a data set consisting of the values $a_{1},a_{2},...,a_{n}$, the average is defined as follows:
$$
\frac{1}{n}\sum_{i=1}^{n}a_{i}
$$



## Example

Consider the data sample ${1,2,3,4}$. The average is defined as follows:

$$
\frac{1+2+3+4}{4}=2.5
$$
